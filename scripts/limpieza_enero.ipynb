```python
import pandas as pd

ruta = 'C:/Users/Sistemas/Desktop/proyecto_pallets/data/enero_pallets.xlsx'

df = pd.read_excel (ruta)

df.head()
```

```python
df.rename(columns={'Unnamed: 0': 'id_socio', 'Unnamed: 1': 'socio'}, inplace=True)
```

```python
df_melted = pd.melt(df, 
                    id_vars=['id_socio', 'socio'], 
                    value_vars=df.columns[2:26], 
                    var_name='fecha', 
                    value_name='pallets')

# Convertir la columna 'fecha' a tipo datetime
df_melted['fecha'] = pd.to_datetime(df_melted['fecha'])

df_melted.head()
```

```python
df_melted.to_excel('df_melted_exportado.xlsx', index=False)
```

```python
# Realizar el merge para asociar las columnas adicionales
df_enriched = pd.merge(df_melted, df[['id_socio', 'socio', 'base', 'Promedio pallets por socio diario', 
                                      'Porcentaje pallets por socio', 'Parte variable', 
                                      'Total final por socio', 'Exceso Palets contra (Fijo+Variable)', 
                                      'Exceso contra Fijo']], 
                       on=['id_socio', 'socio'], 
                       how='left')

df_enriched.head()
```

```python
df_enriched.to_excel('df_melted_exportado.xlsx', index=False)
```

```python
df_enriched['Porcentaje pallets por socio'] = df_enriched['Porcentaje pallets por socio'] * 100
df_enriched['Porcentaje pallets por socio'] = df_enriched['Porcentaje pallets por socio'].map('{:.2f}%'.format)
df_enriched.head()
```

```python
df_enriched['Parte variable'] = df_enriched['Parte variable'].astype(int)
df_enriched.head()
```

```python
df_enriched['Total final por socio'] = df_enriched['Total final por socio'].astype(int)
df_enriched.head()
```

```python
df_enriched['Exceso Palets contra (Fijo+Variable)'] = df_enriched['Exceso Palets contra (Fijo+Variable)'].astype(int)
df_enriched.head()
```

```python
df_enriched.to_excel('df_melted_exportado.xlsx', index=False)
```
